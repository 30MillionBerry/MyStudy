# 머신러닝 모델링이란

머신러닝 모델링.JPG

X와 Y를 준다. X와 Y의 관계를 설명하는 함수식을 찾는게 목적
X와 Y의 관계를 잘 찾는 방법론들, 알고리듬을 인간이 개발하고 컴퓨터에게 알려준다면 
우리에게 해답으로 돌아오는 프로세스가 머신러닝
우리가 찾은 함수가 세탁기에 들어가면 인공지능 세탁기 , 냉장고에 들어가면 인공지능 냉장고

빅데이터모델링 예측.JPG

X와 Y값의 관계를 잘 설명하는 함수를 찾으면 그 함수로 할 수 있는 일들이 많다.
그 중에 가장 대표적인게 예측이다.
함수식을 찾았다. 만약에 Y값이 없는 어떤 X값이 왔을 때 이를 우리가 만들어 놓은 함수에 X값을 집어넣으면 이 함수가 Y값을 알려주는 것이다.  이 결과를 가지고 우리가 예측값으로 활용을 하게되는 것이다.
